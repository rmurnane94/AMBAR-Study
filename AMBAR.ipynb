```python
import pandas as pd
import math
import statistics
import scipy
```

```python
df = pd.read_excel('Book1.xlsx', sheet_name= 'ADAS-Cog', header=[0,1])
# print(df)
```

```python
df.columns = ['.'.join(col).strip() for col in df.columns.values]
df = df.rename(columns={'Unnamed: 0_level_0.Unnamed: 0_level_1': 'Time'})
df.head(10)
```

```python
mean1 = df['Placebo.mean']
std1 = df['Placebo.SD']
nobs1 = df['Placebo.N']


treatments = ['Low Albumin', 'Low Albumin + IVIG', 'High Albumin + IVIG', 'All PE-treated']

for treatment in treatments:
  mean2 = df[treatment + '.mean']
  std2 = df[treatment + '.SD']
  nobs2 = df[treatment + '.N']

  stats = []
  pvalues = []

  for x in range(len(df)):
  # print(x)
  # print(mean1[x]
    stat, pvalue = scipy.stats.ttest_ind_from_stats(mean1[x], std1[x], nobs1[x], mean2[x], std2[x], nobs2[x], equal_var=True, alternative='two-sided')
    # print(treatment, x, stat, pvalue)
    stats.append(round(stat,3))
    pvalues.append(round(pvalue,3))

  df[treatment + ' t_stats'] = stats
  df[treatment + ' p_values'] = pvalues
```
